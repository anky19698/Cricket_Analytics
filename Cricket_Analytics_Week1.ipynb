###  Is MS Dhoni still good enough to bat?

```python
# We will Compare MS Dhoni's IPL Peformance Before and After His International Retirement

# Till  2019 vs After 2019

# Based on Results We can Decide whether MS Dhoni is still good enough to bat or  not
```

```python
msd_till_2019 = msd_df[msd_df['year'] <= 2019]

msd_till_2019.head()
```

```python
msd_after_2019 = msd_df[msd_df['year'] > 2019]

msd_after_2019.head()
```

### Calculating MS Dhoni's Stats:
1) runs scored
2) balls faced
3) strike rate
4) average

```python
runs_till_2019 = sum(list(msd_till_2019['runs_off_bat']))

runs_till_2019
```

```python
balls_faced_till_2019 = len(msd_till_2019)

balls_faced_till_2019
```

```python
strike_rate_till_2019 = 100*runs_till_2019/balls_faced_till_2019

strike_rate_till_2019
```

```python
runs_per_match_till_2019 = list(msd_till_2019.groupby('match_id')['runs_off_bat'].sum())

average_till_2019 = sum(runs_per_match_till_2019)/len(runs_per_match_till_2019)

average_till_2019
```

```python
runs_after_2019 = sum(list(msd_after_2019['runs_off_bat']))

runs_after_2019
```

```python
balls_faced_after_2019 = len(msd_after_2019)

balls_faced_after_2019
```

```python
strike_rate_after_2019 = 100*runs_after_2019/balls_faced_after_2019

strike_rate_after_2019
```

```python
runs_per_match_after_2019 = list(msd_after_2019.groupby('match_id')['runs_off_bat'].sum())

average_after_2019 = sum(runs_per_match_after_2019)/len(runs_per_match_after_2019)

average_after_2019
```

### Top Batsmen in IPL after 2019


```python
players_after_2019 = df[df['year'] > 2019]

players_after_2019.head()
```

```python
players = pd.DataFrame(players_after_2019[['striker', 'runs_off_bat']].groupby('striker')['runs_off_bat'].sum())

balls_faced = pd.DataFrame(players_after_2019[['striker', 'balls_in_over']].groupby('striker')['balls_in_over'].count())

balls_faced



players_df = pd.merge( players, balls_faced, on='striker', how='left')

players_df['strike_rate'] = 100*players_df['runs_off_bat']/players_df['balls_in_over']

players_df


top_batters = players_df[players_df['runs_off_bat'] >= 500 ]

top_batters.reset_index(inplace=True)

top_batters

```

### Batters With Atleast 500 Runs after 2019

```python
import matplotlib.pyplot as plt



# top_batters

plt.scatter(top_batters['strike_rate'] , top_batters['runs_off_bat'])

for i in range(len(top_batters)):
  if(top_batters['striker'][i] in ['KL Rahul', 'Shubman Gill', 'V Kohli', 'MS Dhoni', 'RG Sharma', 'SA Yadav', 'DA Warner', 'KS Williamson']):
    plt.text(top_batters['strike_rate'][i], top_batters['runs_off_bat'][i], top_batters['striker'][i])

plt.axvline(125)
plt.axhline(1400)
plt.xlabel('Strike Rate')
plt.ylabel('Runs Scored')
plt.title('Batters With Atleast 500 Runs after IPL 2019')
plt.show()
```
