## Let us rewind to 8am on Sunday morning, 19th November. Australia have won the toss in the WorldCup Final vs India, and both starting XIs have been announced.

1. At that point in time, (i.e. post-toss, but before a ball has been bowled) what Dream11 Fantasy Team would you have picked for the final, and why? (assume you have infinite budget)
2. At that point in time how many Dream11 fantasy points would you expect each player to score, and why? Can you rank the 22 players from highest to lowest in expected Dream11 fantasy points?

## Solution:

### Approach

1. We will Rank the 22 Players Based on Fantasy Points they collected Before the World Cup Final
2. From Those We will Choose Best Possible XI Based on Rankings and Selection Criteria => WK(1-4), BAT(3-6), BOWL(3-6), AR(1-4)

### Fantasy Data

Since We Don't Have Fantasy Data Available till the World Cup Final, We will Calculate it using Dataset from Cricsheet

We Will Focus on Batting and Bowling Points Only, Since they have Major impact on Points system and Fielding Data isn't available in Dataset.

```python
# Get the ODI WC Ball by Ball Data

import requests
import zipfile
import io
import os
import pandas as pd
import glob
import os
import numpy as np

url = 'https://cricsheet.org/downloads/icc_mens_cricket_world_cup_csv2.zip'

response = requests.get(url)
destination_folder = 'odi_wc_male_csv'

if response.status_code == 200:
    with zipfile.ZipFile(io.BytesIO(response.content), 'r') as zip_ref:
        zip_ref.extractall(destination_folder)
        print('success')
else:
    print('failed')


```

```python
files = glob.glob('/content/odi_wc_male_csv/[0-9]*.csv')
all_files = []
for file in files:
  filename = os.path.basename(file)
  # print(filename[:-4])
  if 'info' not in filename:
    # print(filename)
    all_files.append(file)

df = pd.concat((pd.read_csv(f, header = 0) for f in all_files))

# exporting final csv without super overs data

df = df[df['innings'] < 3]

df.to_csv('odi_wc_ball_by_ball.csv', index=False)
```

## Batting Points

```python
runs = df.groupby(['match_id', 'striker'])['runs_off_bat'].sum().reset_index().rename(columns = {'runs_off_bat': 'runs'})
balls_faced = df.groupby(['match_id', 'striker'])['ball'].count().reset_index().rename(columns = {'ball': 'balls_faced'})
sixes = df.groupby(['match_id', 'striker'])['is_six'].sum().reset_index().rename(columns = {'is_six': 'sixes'})
fours = df.groupby(['match_id', 'striker'])['is_four'].sum().reset_index().rename(columns = {'is_four': 'fours'})


runs, balls_faced, sixes, fours
```

```python
batting = pd.merge(runs, balls_faced, on=['match_id','striker']).merge(sixes, on=['match_id','striker']).merge(fours, on=['match_id','striker'])
```

```python
batting.head()
```

```python
batting['fifty'] = batting['runs'].apply(lambda x: 1 if x>=50 and x<100 else 0)
batting['hundred'] = batting['runs'].apply(lambda x: 1 if x>=100 else 0)
```

```python
batting['points'] = batting['runs'] + batting['fours'] + batting['sixes'] * 2 + batting['fifty'] * 4 + batting['hundred'] * 8
batting.head()
```

## Bowling Points

```python
df.columns
```

```python
df.wicket_type.unique()
```

```python
not_bowlers_wkt = ['run out', 'retired hurt']
```

```python
def is_bowlers_wkt(dismissal_kind):

  if dismissal_kind not in not_bowlers_wkt and type(dismissal_kind) == str:
    return 1
  else:
    return 0

def is_bonus_wkt(dismissal_kind):

  if dismissal_kind in ['bowled', 'lbw'] and type(dismissal_kind) == str:
    return 1
  else:
    return 0
```

```python
df['is_bowlers_wkt'] = df.apply(lambda x: is_bowlers_wkt(x['wicket_type']), axis=1)
df['is_bonus_wkt'] = df.apply(lambda x: is_bonus_wkt(x['wicket_type']), axis=1)
df[['wicket_type', 'is_bowlers_wkt', 'is_bonus_wkt']].head(20)

```

```python
df['over'] = df['ball'].apply(lambda x: int(str(x).split('.')[0]))
df['total_runs'] = df['runs_off_bat'] + df['extras']
```

```python
df.head()
```

```python
wkt = df.groupby(['match_id', 'bowler'])['is_bowlers_wkt'].sum().reset_index()
bonus_wkt = df.groupby(['match_id', 'bowler'])['is_bonus_wkt'].sum().reset_index()
```

```python
overs_balled = df.groupby(['match_id', 'bowler', 'over'])['total_runs'].sum().reset_index()
overs_balled['maiden'] = overs_balled['total_runs'].apply(lambda x: 1 if x==0 else 0)

overs_balled.head()

maidens = overs_balled.groupby(['match_id', 'bowler'])['maiden'].sum().reset_index()

maidens
```

```python
bowling = pd.merge(wkt, bonus_wkt, on=['match_id','bowler']).merge(maidens, on=['match_id','bowler'])
bowling.head()
```

```python
bowling['4w'] = bowling['is_bowlers_wkt'].apply(lambda x: 1 if x>=4 else 0)
bowling['5w'] = bowling['is_bowlers_wkt'].apply(lambda x: 1 if x>=5 else 0)
```

```python
bowling.head()
```

```python
bowling['points'] = bowling['is_bowlers_wkt'] * 25 + bowling['is_bonus_wkt'] * 8 + bowling['4w'] * 4 + bowling['5w'] * 8 + bowling['maiden'] * 4
```

```python
bowling.head()
```

## Calculating Total Fantasy Points

```python
bowling = bowling.rename(columns= {'bowler': 'player'})
batting = batting.rename(columns= {'striker': 'player'})
```

```python
player = pd.merge(batting, bowling, on=['match_id','player'], how='outer')
```

```python
player = player.fillna(0)
player.head()
```

```python
player['total_points'] = player['points_x'] + player['points_y']
```

```python
final_players = player[player['player'].isin(cricket_players)]

fantasy_points = final_players.groupby(['player'])['total_points'].sum().reset_index()
```
